```python
import numpy as np
import matplotlib.pyplot as pl

page_speed = np.random.normal(6,1.3,10000)
purchase_amount = np.random.normal(5,0.3,10000)/page_speed *100
pl.scatter(page_speed,purchase_amount)
```

```python
x = np.array(page_speed)
y = np.array(purchase_amount)
pol = np.poly1d(np.polyfit(x,y,8))
```

```python
line = np.linspace(0,13,1000)
pl.scatter(page_speed,purchase_amount)
pl.plot(line , pol(line),c='r')
pl.show()
```

```python
from sklearn.metrics import r2_score
r2_percent = r2_score(y,pol(x))
r2_percent
```
